### Проверка расчета эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака `rougher.output.recovery` по следующей формуле:

***Recovery*** = ${C*(F-T)\over F*(C-T)}$*100% , где

* **C** - доля золота в концентрате после флотации; 
* **F** - доля золота в сырье до флотации;
* **T** - доля золота в отвальных хвостах после флотации.

Далее найдем MAE между произведенными расчётами и значением признака в исходных данных (в выборке `train_sample`):

***MAE*** = $1\over N$$\sum\limits_{i=1}^{N}$${abs(y_i - \overline{y}_i)} $ , где

* **$y_i$** - значение признака, полученное из исходных данных; 
* **$\overline{y}_i$** - рассчитанное значение признака.

```python
def mae(row):
    c = row['rougher.output.concentrate_au']
    f = row['rougher.input.feed_au']
    t = row['rougher.output.tail_au']
    
    recovery = row['rougher.output.recovery']
    result = (c * (f-t))/(f * (c-t)) * 100
    return abs(result-recovery)

mae = (train_sample.apply(mae, axis = 1)).sum()/train_sample['rougher.output.recovery'].count()
print('Средняя абсолютная ошибка:', mae)
```

Полученное значение средней абсолютной ошибки достаточное небольшое, что говорит о том, что значения эффективности обогащения, представленные в качестве исходных данных, были рассчитанны корректно.

### Изменение концентрации металлов

Определим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки.

```python
def conc_plot(stages):
    plt.figure(figsize=(14,5))
  
    for stage in stages:
        sns.distplot(data_new[stage], hist=True, kde=True, bins=80)
        
    plt.legend(stages)
    plt.xlabel(r'Концентрация')
    plt.ylabel(r'Частота параметра')
    plt.title(r'Изменение концентрации на различных этапах очистки')
     
def feed_tail(conc, tail):
    x = ['До флотации', 'Флотация', 'Первичная очистка', 'Вторичная очистка']
    x1 = ['Флотация', 'Первичная очистка', 'Вторичная очистка']
    plt.figure(figsize=(7,4))
    plt.plot(x, conc, '--o')
    plt.plot(x1, tail, '--o')
    plt.grid()
    plt.legend(['Концентрация в сырье', 'Концентрация в хвостах'])
```

* **Золото**

```python
stages_au = ['rougher.input.feed_au', 'rougher.output.concentrate_au', 'primary_cleaner.output.concentrate_au', 
             'final.output.concentrate_au']

conc_plot(stages_au)
plt.xlim([0, 50])
```

Как видно, с течением времени постепенно увеличивается концентрация золота в смеси золотосодержащей руды/концентрата - пики на гистограммах на этапах очистки постепенно смещаются вправо. Максимальное медианное значение концентрации соответсвует финальному концентрату после этапа вторичной очистки - около 45.

Ближе к 0 видны небольшие пики - это можно объяснить тем, что нулевая концентрация золота могла регистрироваться в те моменты времени, когда смесь еще не дошла до определенного этапа очистки. 

Посмотрим, как меняется концентрация золота в сырье и в отвальных хвостах на разных этапах очистки.

```python
conc_au = [data_new['rougher.input.feed_au'].mean(),
      data_new['rougher.output.concentrate_au'].mean(),
      data_new['primary_cleaner.output.concentrate_au'].mean(),
      data_new['final.output.concentrate_au'].mean()]

tail_au = [data_new['rougher.output.tail_au'].mean(),
      data_new['primary_cleaner.output.tail_au'].mean(),
      data_new['final.output.tail_au'].mean()]

feed_tail(conc_au, tail_au)
```

В процессе получения золота из руды его количество в концентрате ожидаемо увеличивается. Небольшое количество золота в связи с несовершенством технологического процесса на различных этапах очистки удаляется из концентрата вместе с отвальными хвостами, но в очень небольшом количестве.

###  Суммарная концентрация всех веществ на разных стадиях

Исследуем суммарную концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах.

```python
data_new['rougher.input_conc'] = data_new['rougher.input.feed_au'] + data_new['rougher.input.feed_ag']+data_new['rougher.input.feed_pb']+ data_new['rougher.input.feed_sol']
data_new['rougher.output_conc'] = data_new['rougher.output.concentrate_au'] + data_new['rougher.output.concentrate_ag'] + data_new['rougher.output.concentrate_sol'] 
data_new['primary_conc'] = data_new['primary_cleaner.output.concentrate_au'] + data_new['primary_cleaner.output.concentrate_ag'] + data_new['primary_cleaner.output.concentrate_pb'] + data_new['primary_cleaner.output.concentrate_sol']
data_new['final_conc'] = data_new['final.output.concentrate_au'] + data_new['final.output.concentrate_ag'] + data_new['final.output.concentrate_pb'] + data_new['final.output.concentrate_sol']  

def sum_conc_plot(col):
    sns.distplot(col, hist=True, kde=True, bins=80)  
    plt.xlim([-1, 85])
    plt.ylim([-0.001,0.17])
    plt.title(r'Суммарная концентрация всех веществ в сырье/концентрате')
    plt.xlabel(r'Cуммарная концентрация')
    plt.ylabel(r'Частота параметра')
```

Получаем, что минимальная медианная суммарная концентрация соответсвует 1 этапу (исходное сырье), максимальная - последнему (финальный концентрат). Это связано с тем, что в финальном концентрате уже удалено максимальное количество посторонних ненужных веществ, поэтому концентрация золота, серебра, свинца и солей в сокращенном объеме смеси дает большее значение концентрации.

Слева также наблюдаются пики рядом с нулем. Суммарная концентрация веществ не может быть нулевой. Такие значения могли появиться вследствие технической ошибки, когда концентрации веществ временно переставали регистрироваться. 

При обучении модели такие данные будут неинформативными, поэтому их можно удалить из обучающей и тестовой выборок.

Выделим из общего датасета даты, которым соответствуют нулывые значения суммарных концентраций.

```python
def def_dates(row):
    rougher_input_conc = row['rougher.input_conc'] 
    rougher_output_conc = row['rougher.output_conc'] 
    primary_conc = row['primary_conc']
    final_conc = row['final_conc']
    date = row['date']
    res = None
    if (rougher_input_conc == 0) or (rougher_output_conc == 0) or (primary_conc == 0) or (final_conc == 0):
        res = date
    return res
    
dates = []
dates = data_new.apply(def_dates, axis = 1).dropna(axis=0)

display(dates.count())
```

## Модель

Перед обучением и тестированием модели выделим общие и целевые признаки для обучающей и тестовой выборок и напишем функцию по вычислению метрики sMAPE для оценки качества модели.

```python
# Преобразуем столбец с датами и временем в индексы строк
train_sample_new = train_sample_new.set_index('date')
test_sample_new = test_sample_new.set_index('date')

# Выделяем общие и целевые признаки
# Для обучающей выборки
target_rougher_train =  pd.Series(train_sample_new['rougher.output.recovery'])
target_final_train =  pd.Series(train_sample_new['final.output.recovery'])
features_train = train_sample_new.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)

# Для тестовой выборки
target_rougher_test = pd.Series(test_sample_new['rougher.output.recovery'])
target_final_test =  pd.Series(test_sample_new['final.output.recovery'])
features_test = test_sample_new.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)
```

Напишем функцию для вычисления метрики sMAPE.


```python
def smape(target, prediction, **kwargs):
    score = 100/len(target) * np.sum(2 * np.abs(target - prediction) / (np.abs(target) + np.abs(prediction)))
    return score
```

### Выбор модели

Для наших целей обучим две модели: *Решающего дерева* и *Линейной регрессии*. Оценим их качество кросс-валидацией и выберем лучшую модель.

```python
def model_prep(model, parameters, name):
    # Стандартизация данных
    scaler =-()
    steps = [('normalizer', scaler), (name, model)]
    pipe = Pipeline(steps)
    # Гипермараметры для решающего дерева
    params = parameters
    # Функция для расчета метрики sMAPE преобразуется в скорер
    smape_scorer = make_scorer(smape, greater_is_better=False)
    grid = GridSearchCV(estimator=pipe, param_grid=params, scoring=smape_scorer, cv=3)
    return grid
```

* **Решающее дерево**

```python
# Предсказание rougher.output.recovery

# Обучение модели решающего дерева
rougher_tree = model_prep(model = DecisionTreeRegressor(random_state=12345),
    parameters={'rougher_tree__max_depth': np.arange(1, 21),
               'rougher_tree__min_samples_leaf': [1, 5, 10, 20, 50, 100]},
    name='rougher_tree')

rougher_tree.fit(features_train, target_rougher_train)

# Получение предсказаний
predictions_test_rougher_tree = rougher_tree.predict(features_test)

# Оценка качества модели
smape_rougher_tree = smape(target_rougher_test, predictions_test_rougher_tree)

print('Значение sMAPE для rougher.output.recovery на модели решающего дерева: {:.4f}%'.format(smape_rougher_tree))
```

```python
# Предсказание final.output.recovery

# Обучение модели решающего дерева
final_tree = model_prep(model = DecisionTreeRegressor(random_state=12345),
    parameters={'final_tree__max_depth': np.arange(1, 21),
               'final_tree__min_samples_leaf': [1, 5, 10, 20, 50, 100]},
    name='final_tree')

final_tree.fit(features_train, target_final_train)

# Получение предсказаний
predictions_test_final_tree = final_tree.predict(features_test)

# Оценка качества модели
smape_final_tree = smape(target_final_test, predictions_test_final_tree)

print('Значение sMAPE для final.output.recovery на модели решающего дерева: {:.4f}%'.format(smape_final_tree))
```

* **Линейная регрессия**

```python
# Предсказание rougher.output.recovery

# Обучение модели линейной регрессии
rougher_linear = model_prep(model=LinearRegression(),
    parameters={'rougher_linear__fit_intercept': [True, False],
                'rougher_linear__copy_X': [True, False]},
    name='rougher_linear')

rougher_linear.fit(features_train, target_rougher_train)

# Получение предсказаний
predictions_test_rougher_linear = rougher_linear.predict(features_test)

# Оценка качества модели
smape_rougher_linear = smape(target_rougher_test, predictions_test_rougher_linear)

print('Значение sMAPE для rougher.output.recovery на модели линейной регрессии: {:.4f}%'.format(smape_rougher_linear))
```

```python
# Предсказание final.output.recovery

# Обучение модели линейной регрессии
final_linear = model_prep(model=LinearRegression(),
    parameters={'final_linear__fit_intercept': [True, False],
                'final_linear__copy_X': [True, False]},
    name='final_linear')

final_linear.fit(features_train, target_final_train)

# Получение предсказаний
predictions_test_final_linear = final_linear.predict(features_test)

# Оценка качества модели
smape_final_linear = smape(target_final_test, predictions_test_final_linear)

print('Значение sMAPE для final.output.recovery на модели линейной регрессии: {:.4f}%'.format(smape_final_linear))
```

Минимальные значения sMAPE были достигнуты для модели решающего дерева: для признака `rougher.output.recovery` - 4.92%, для `final.output.recovery` - 8.02%. 

Поэтому выбираем модель решающего дерева.

### Проверка модели на адекватность

При создании модели машинного обучения для какого-либо проекта разумно создать baseline модель, представляющую собой простейшую (*dummy*) модель, которая всегда предсказывает наиболее часто встречающийся класс. Это позволяет получить контрольные данные для сравнительной оценки построенной нами ранее более сложной модели и быть уверенными в качестве ее работы (что она выдает не просто набор случайно подобранных данных).

Для проверки модели на адекватность обучим модель *Dummy* для нашей задачи на обучающей выборке и применим ее к тестовым данным. 

```python
def dummy(features_train, features_test, target_train, target_test):
    dummy = DummyRegressor(strategy='mean').fit(features_train, target_train)
    predictions_dummy = dummy.predict(features_test)
    smape_dummy = smape(target_test, predictions_dummy)
    return smape_dummy

# Для признака rougher.output.recovery
smape_rougher_dummy = dummy(features_train, features_test, target_rougher_train, target_rougher_test)
print('Значение sMAPE для rougher.output.recovery: {:.4f}%'.format(smape_rougher_dummy))

# Для признака final.output.recovery
smape_final_dummy = dummy(features_train, features_test, target_final_train, target_final_test)
print('Значение sMAPE для final.output.recovery: {:.4f}%'.format(smape_final_dummy))
```

При проверке модели было получено значение sMAPE 5.33% для признака `для rougher.output.recovery` и 8.53% для признака `final.output.recovery`, что больше ранее полученных значений метрик на выбранной модели, что позволяет сделать вывод о возможности применимости обученной нами модели решающего дерева.
